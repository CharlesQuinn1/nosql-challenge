### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':{'$in':[20]}}

# Use count_documents to display the number of documents in the result
local_auth = establishments.count_documents(query)
print(f"There are {local_auth} establishments with a hygiene score of 20.")

# Display the first document in the results using pprint
establishments.find_one(query)
```

```python
pprint(local_auth)
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
hygiene_20_df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
print(f"There are {len(hygiene_20_df)} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
hygiene_20_df.iloc[:10,:]
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'$and': [{'LocalAuthorityName':{'$in':['City of London Corporation']}},{'RatingValue':{'$gte':4}}]}

# Use count_documents to display the number of documents in the result
result_count = establishments.count_documents(query)
print(f"There are {result_count} establishments London as the Local Authority and has a RatingValue greater than or equal to 4.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
results_df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
count_results_df = len(results_df)
print(f"There are {count_results_df} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
results_df.iloc[:10,:]
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match':{'scores.Hygiene':{'$eq':0}}}
query = {'$group':{'_id': '$LocalAuthorityName', 'count':{'$sum':1}}}
sort_query = {'$sort': {'count':-1}}
pipeline = [match_query,query,sort_query]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
total = 0
for x in results:
    total += x['count']

print(f"There are {len(results)} documents in the result with a total document count of {total}.")

# Print the first 10 results
print(results[:10])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
print(f"There are {len(results_df)} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
results_df.iloc[:10,:]
```
